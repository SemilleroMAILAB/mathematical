</center>
 <H3>
Objective
</H3>
<P>
The goal of this course is to study the elements that support to student understand the computing nature applien on pattern recognition. The study of automatas, gramtics and formal languages allow to engineer of system to build mathematical models in software design, such as: language processing, data base search, web searching, etc. The student must know basics mathematics,discrete mathematics, one-variables calculus, computational logic and computer programming.
</P>
