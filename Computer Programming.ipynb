Course Description
</H2>
</center>
 <H3>
Objective
</H3>
<P>
The goal of this course is to study the elements that support to student realize  the activities used for computer programming and logical mathematical. The student must know basics mathematics and arithmetic.
</P>
<H3>
Methodology
</H3>
<ul>
<li>Professor's master classes.</li>
<li>Question and dudes solving by students and teacher.</li>
<li><b>Programming languages</b>
<ul>
<li>Visua Basic Language</li>
</ul>
</li>
<li>Individual-Exam, Two-person exam, Home Exam.</li>
<li>Final Project!</li>

</ul>
