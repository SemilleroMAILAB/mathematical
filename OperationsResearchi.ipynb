 <H3>
Objective
</H3>
<P>
The goal of this course is to study the elements that support to student realize  the activities used for mathematical and deterministic modelling, linear and non-linear programming applied to real life. The student must know basics mathematics, data structures, computer programming, differential equations, statistics, linear algebra, discrete mathematics, multi-variable calculus, physics mechanics, geometry, analysis and design algorithms.
</P>
